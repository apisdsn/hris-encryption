```python
import matplotlib.pyplot as plt
import json

# Example JSON data
data = [
    {"metode_http": "get", "metode_enkripsi": "asymmetric", "jumlah_data": 10, "algoritma": "aes256", "response_time_get": "40658 ms"},
    {"metode_http": "insert", "metode_enkripsi": "asymmetric", "response_time_insert": "402 ms", "jumlah_data": 10, "algoritma": "aes256"},
    {"metode_http": "get", "metode_enkripsi": "asymmetric", "jumlah_data": 50, "algoritma": "aes256", "response_time_get": "80658 ms"},
    {"metode_http": "insert", "metode_enkripsi": "asymmetric", "response_time_insert": "602 ms", "jumlah_data": 50, "algoritma": "aes256"},
    {"metode_http": "get", "metode_enkripsi": "asymmetric", "jumlah_data": 10, "algoritma": "blowfish", "response_time_get": "20658 ms"},
    {"metode_http": "insert", "metode_enkripsi": "asymmetric", "response_time_insert": "132 ms", "jumlah_data": 10, "algoritma": "blowfish"},
    {"metode_http": "get", "metode_enkripsi": "asymmetric", "jumlah_data": 50, "algoritma": "blowfish", "response_time_get": "40658 ms"},
    {"metode_http": "insert", "metode_enkripsi": "asymmetric", "response_time_insert": "702 ms", "jumlah_data": 50, "algoritma": "blowfish"}
]

# Prepare data for plotting
def parse_time(time_str):
    return int(time_str.replace(' ms', '').strip())

def extract_data(data):
    response_times = {'get': {'aes256': [], 'blowfish': []}, 'insert': {'aes256': [], 'blowfish': []}}
    data_sizes = {'get': {'aes256': [], 'blowfish': []}, 'insert': {'aes256': [], 'blowfish': []}}
    
    for entry in data:
        method = entry['metode_http']
        algorithm = entry['algoritma']
        size = entry['jumlah_data']
        
        if method == 'get':
            response_times['get'][algorithm].append(parse_time(entry['response_time_get']))
            data_sizes['get'][algorithm].append(size)
        elif method == 'insert':
            response_times['insert'][algorithm].append(parse_time(entry['response_time_insert']))
            data_sizes['insert'][algorithm].append(size)
    
    return response_times, data_sizes

response_times, data_sizes = extract_data(data)

# Plotting
def plot_line_graph(data_sizes, response_times):
    plt.figure(figsize=(14, 8))
    
    for algorithm in ['aes256', 'blowfish']:
        plt.plot(data_sizes['get'][algorithm], response_times['get'][algorithm], marker='o', label=f'pgp_pub_decrypt() GET - {algorithm}')
        plt.plot(data_sizes['insert'][algorithm], response_times['insert'][algorithm], marker='o', linestyle='--', label=f'pgp_pub_encrypt() INSERT - {algorithm}')
    
    plt.xlabel('Jumlah Data')
    plt.ylabel('Response Time (ms)')
    plt.title('Asymmetric Encryption Response Time vs. Data Size')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_line_graph(data_sizes, response_times)
```
